```python
import matplotlib as pt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Oumaima Hoummadi - efficacite_energetique_Dataset.csv")
```

```python
df.head()
```

```python
X=df.drop("Energy_Efficiency_Rating",axis=1)
y=df["Energy_Efficiency_Rating"]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_train)
y_pred
```

```python
acc_train=r2_score(y_pred,y_train)
```

```python
y_pred_test=model.predict(X_test)
acc_test=r2_score(y_pred_test,y_test)
```

```python
(acc_train,acc_test)
# 9%
```

```python
from sklearn.linear_model import LassoCV
lambda1 = np.linspace(0, 10, 20)  
lassoCV = LassoCV(alphas=lambda1, cv=10, max_iter=1000000)
lassoCV.fit(X_train,y_train)
best_alpha = lassoCV.alpha_
print("best vel alpha ",best_alpha)
```

```python
coef = pd.DataFrame({
    "feature": X_train.columns,
    "avant L1": model.coef_,
    "après L1": lassoCV.coef_
})

coef
```

```python
from sklearn.linear_model import Lasso
reg_L1 = Lasso(alpha=best_alpha)
reg_L1.fit(X_train, y_train)
```

```python
y_pred_train_l1=reg_L1.predict(X_train)
y_pred_test_l1=reg_L1.predict(X_test)
```

```python
acc_train_L1=r2_score(y_pred_train_l1,y_train)
acc_test_L1=r2_score(y_pred_test_l1,y_test)
```

```python
(acc_train_L1,acc_test_L1)
# 7%
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

class DataProcessor:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None
        self.original_shape = None

    def load_data(self):
        print("Step 1: Loading and examining data...")
        
        # Load the data
        self.df = pd.read_csv(self.file_path)
        self.original_shape = self.df.shape
        
        print("\nDataset Overview:")
        print(f"Shape: {self.df.shape}")
        print("\nColumns:", self.df.columns.tolist())
        print("\nMissing Values:\n", self.df.isnull().sum())
        
        return self

    def check_data_quality(self):
        """
        Step 2: Check data quality
        """
        print("\nStep 2: Checking data quality...")
        
        # Check for duplicates
        duplicates = self.df.duplicated().sum()
        print(f"Number of duplicate rows: {duplicates}")
        
        # Check data types
        print("\nData Types:\n", self.df.dtypes)
        
        # Check basic statistics
        print("\nNumerical Columns Statistics:")
        print(self.df.describe())
        
        return self

    def clean_data(self):
        print("\nStep 3: Cleaning data...")
        
        # Remove duplicates if any
        self.df = self.df.drop_duplicates()
        
        # Handle missing values if any
        if self.df.isnull().sum().any():
            # Fill numerical missing values with median
            numerical_columns = self.df.select_dtypes(include=[np.number]).columns
            self.df[numerical_columns] = self.df[numerical_columns].fillna(self.df[numerical_columns].median())
        
        # Remove outliers using IQR method for numerical columns
        for col in self.df.select_dtypes(include=[np.number]).columns:
            Q1 = self.df[col].quantile(0.25)
            Q3 = self.df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            self.df[col] = np.where(self.df[col] > upper_bound, upper_bound,
                                  np.where(self.df[col] < lower_bound, lower_bound, self.df[col]))
        
        return self

    def generate_report(self):
        print("\nStep 8: Generating final report...")
        
        print("\nFeature Engineering Summary:")
        print(f"Original number of features: {self.original_shape[1]}")
        print(f"Final number of features: {self.df.shape[1]}")
        print(f"\nNew features created:")
        new_features = [col for col in self.df.columns if col not in pd.read_csv(self.file_path).columns]
        for feature in new_features:
            print(f"- {feature}")
        
        return self
    
    def scale_features(self):
        """
        Step 7: Scale features
        """
        print("\nStep 7: Scaling features...")
        
        # Identify numerical columns to scale
        numerical_cols = ['Wall_Area', 'Roof_Area', 'Window_Area', 'Overall_Height',
                         'Outdoor_Temperature', 'Humidity', 'Total_Surface_Area',
                         'Window_to_Wall_Ratio', 'Roof_to_Wall_Ratio', 'Temp_Humidity_Interaction']
        
        # Scale numerical features
        scaler = StandardScaler()
        self.df[numerical_cols] = scaler.fit_transform(self.df[numerical_cols])
        
        return self

    def process_all(self):
        (self.load_data()
         .check_data_quality()
         .clean_data()
         .scale_features()
         .generate_report())
        
        return self.df

# Example usage
if __name__ == "__main__":
    # Replace with your CSV file path
    file_path = "Mohamed Belannab - efficacite_energetique_Dataset.csv"
    
    try:
        # Create processor and run all steps
        processor = DataProcessor(file_path)
        processed_df = processor.process_all()
        
        print("\nProcessing completed successfully!")
        
    except Exception as e:
        print(f"Error during processing: {str(e)}")
```
