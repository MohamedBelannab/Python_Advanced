```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(w):
    return np.sin(w[0]) + (w[1] + 3)**2 + np.exp(0.5*w[0])

def grad_f(w) : 
    df_dw1 = np.cos(w[0]) + 0.5*np.exp(0.5*w[0])
    df_w2 = 2*(w[1]+ 3)
    return np.array([df_dw1 , df_w2])

def descente_gradient(w , alpha=0.1 , n_iteration=50):
    temp = [w.copy()]
    for i in range(n_iteration):
        w= w-alpha * grad_f(w)
        temp.append(w.copy())
    return np.array(temp)

x1 = np.linspace(-5 , 3 , 100)
x2 = np.linspace(-7 , 1 , 100)
w = np.meshgrid(x1,x2)
z = f(w)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121 , projection='3d')
ax1.plot_surface(w[0] , w[1] , z)
ax2 = fig.add_subplot(121)
ax2.contour(w[0] , w[1] , z , cmap="plasma" , levels=20)
w=np.array([0,0,0,0])
trajectoires = {"descent gradient" : descente_gradient(w)}
print(trajectoires)
```
